<h2>Install pandas and use it to read the .csv file and create a data frame out of it</h2>

```python
#!pip install -U pandas
import pandas as pd

flights = pd.read_csv('formatted_flights.csv')
print(flights.head(10))
```

<h2>Install matplotlib and create histograms with each airline's flights data</h2>

```python
#!pip install -U matplotlib
import matplotlib.pyplot as plt

airlines = flights['name'].unique()
figure, a = plt.subplots(int(airlines.shape[0] / 2), 2, figsize=(12, 24))
a = a.ravel()
x_axes = airlines
y_axes = ['Flights' for x in airlines]

for idx, ax in enumerate(a, start=0):
  is_this_flight = flights['name'] == airlines[idx]
  ax.hist(flights[is_this_flight]['arr_delay'], bins=int(180/5), color='blue', edgecolor='black')
  ax.set_title(airlines[idx])
  ax.set_xlabel('Delay (min)', size=12)
  ax.set_ylabel('Flights', size=12)
  
plt.tight_layout()
plt.show()
```

<h2>For each airline, calculate the chances of arriving on time</h2>

```python
print('Chances of arriving on time with each airline\n')

chances = []
for i in range(len(airlines)):
  this_airline_flights = flights.loc[flights['name'] == airlines[i]]
  this_airline_delays = this_airline_flights['arr_delay']
  on_time_delays = [x for x in this_airline_delays if x <= 0]
  chance = len(on_time_delays) / len(this_airline_delays)
  chance = round(100 * chance, 2)
  chances.append({'airline': airlines[i], 'chance': chance})

sorted_chances = sorted(chances, key=lambda x: x['chance'], reverse=True)
for i, chance in enumerate(sorted_chances):
  print('{0}) {1}: {2}%'.format(i + 1, chance['airline'], chance['chance']))
  
  
```

<h2>In conlusion, according to the available data, Alaska Airlines Inc. is the most likely option to arrive on time.</h2>
